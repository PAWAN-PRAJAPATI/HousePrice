```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
df.head()
```

```python
corrmat = df.corr()
cols = corrmat.nlargest(37,'SalePrice')['SalePrice'].index
zoomcorr = np.corrcoef(df[cols].values.T)
```

```python
X = df[cols]
```

```python
y = X['SalePrice']
X = X.drop(['SalePrice'], axis=1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split

my_imputer = Imputer()
X_ = my_imputer.fit_transform(X)
```

```python
X = pd.DataFrame(X)
train_X, test_X, train_y, test_y = train_test_split(X.as_matrix(), y.as_matrix(), test_size=0.20)
```

```python
from xgboost import XGBRegressor
from sklearn.cross_validation import cross_val_score


my_model = XGBRegressor(n_estimators=1000)
scores = cross_val_score(estimator=my_model,X=train_X,y=train_y,cv=15,n_jobs=-1)
print(scores.mean())
print(scores.std())
```

```python
my_model.fit(train_X,train_y)
my_model.score(test_X,test_y)
```
